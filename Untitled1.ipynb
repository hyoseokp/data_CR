```python
import os
import time
import shutil
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# 0) /
# ==========================
GDS_DIR            = r"D:\dataset\gds"
RAW16_PATH         = r"D:\dataset\raw_input_matrix_1.npy"             # (N,1,16,16) or (N,16,16)
STRUCT128_PATH     = r"D:\dataset\binary_dataset_128_1.npy"           # (N,1,128,128) or (N,128,128)
SPECTRA_PATH       = r"D:\dataset\spectra_result\spectra_latest_1.npy"       # (N,3,301)
VALID_PATH         = r"D:\dataset\spectra_result\spectra_latest_valid_1.npy" # (N,)

SPECTRAL_POINTS = 301
ONLY_VALID1 = True            # True valid==1  idx 
INCLUDE_VALID2 = False        # ONLY_VALID1=False , valid==2 "" 
PRINT_RAW16_NUMBERS = False   # 16x16   

#    (    )
SNAP_DIR = r"D:\dataset\spectra_result\snapshots_plot"
os.makedirs(SNAP_DIR, exist_ok=True)

def norm_path(p: str) -> str:
    return os.path.abspath(p).replace("\\", "/")

GDS_DIR        = norm_path(GDS_DIR)
RAW16_PATH     = norm_path(RAW16_PATH)
STRUCT128_PATH = norm_path(STRUCT128_PATH)
SPECTRA_PATH   = norm_path(SPECTRA_PATH)
VALID_PATH     = norm_path(VALID_PATH)
SNAP_DIR       = norm_path(SNAP_DIR)

# ==========================
# 1)    (  )
# ==========================
def make_snapshot(src_path: str, snap_dir: str, tag: str) -> str:
    """
       npy   ,   read-only .
    - copy2:   
    -          + size  
    """
    src_path = norm_path(src_path)
    ts = time.strftime("%Y%m%d_%H%M%S")
    dst = norm_path(os.path.join(snap_dir, f"{os.path.basename(src_path)}.{tag}.{ts}.snap.npy"))

    def _stat(p):
        st = os.stat(p)
        return st.st_size, st.st_mtime

    #           .
    for attempt in range(1, 6):
        try:
            s1 = _stat(src_path)
            shutil.copy2(src_path, dst)
            time.sleep(0.3)
            s2 = _stat(src_path)

            #     (= ) OK 
            if s1[0] == s2[0]:
                return dst

            #      
            try:
                os.remove(dst)
            except Exception:
                pass
            time.sleep(0.7)
        except Exception as e:
            if attempt == 5:
                raise RuntimeError(f"snapshot failed for {src_path}: {e}") from e
            time.sleep(0.7)

    return dst  #  

raw16_snap   = make_snapshot(RAW16_PATH,     SNAP_DIR, tag="raw16")
struct_snap  = make_snapshot(STRUCT128_PATH, SNAP_DIR, tag="struct128")
spec_snap    = make_snapshot(SPECTRA_PATH,   SNAP_DIR, tag="spectra")
valid_snap   = make_snapshot(VALID_PATH,     SNAP_DIR, tag="valid")

print("snapshots:")
print(" raw16  :", raw16_snap)
print(" struct :", struct_snap)
print(" spectra:", spec_snap)
print(" valid  :", valid_snap)

# ==========================
# 2)  ( read-only)
# ==========================
raw16   = np.load(raw16_snap,  mmap_mode="r")
struct  = np.load(struct_snap, mmap_mode="r")
spec    = np.load(spec_snap,   mmap_mode="r")
valid   = np.load(valid_snap,  mmap_mode="r")

n = min(raw16.shape[0], struct.shape[0], spec.shape[0], valid.shape[0])
print("Common N:", n)
print("raw16  shape/dtype:", raw16.shape, raw16.dtype)
print("struct shape/dtype:", struct.shape, struct.dtype)
print("spec   shape/dtype:", spec.shape, spec.dtype)
print("valid  shape/dtype:", valid.shape, valid.dtype)

# ==========================
# 3) shape sanity ( )
# ==========================
def _check_raw16_shape(a):
    if a.ndim == 3 and a.shape[1:] == (16,16):
        return
    if a.ndim == 4 and a.shape[1:] == (1,16,16):
        return
    raise ValueError(f"raw16 expected (N,16,16) or (N,1,16,16) but got {a.shape}")

def _check_struct128_shape(a):
    if a.ndim == 3 and a.shape[1:] == (128,128):
        return
    if a.ndim == 4 and a.shape[1:] == (1,128,128):
        return
    raise ValueError(f"struct expected (N,128,128) or (N,1,128,128) but got {a.shape}")

_check_raw16_shape(raw16)
_check_struct128_shape(struct)

if spec.ndim != 3 or spec.shape[1] != 3:
    raise ValueError(f"spectra expected (N,3,P) but got {spec.shape}")
if spec.shape[2] != SPECTRAL_POINTS:
    print(f" spectra points mismatch: {spec.shape[2]} (expected {SPECTRAL_POINTS})")

# ==========================
# 4) "  idx" 
# ==========================
v = np.asarray(valid[:n])

if ONLY_VALID1:
    saved_mask = (v == 1)
    label = "valid==1"
else:
    if INCLUDE_VALID2:
        saved_mask = (v == 1) | (v == 2)
        label = "valid in {1,2}"
    else:
        saved_mask = (v != 0)
        label = "valid!=0"

saved_idx = np.where(saved_mask)[0]
if saved_idx.size == 0:
    raise RuntimeError(f"   ({label})")

idx = int(saved_idx.max())
print(f"picked last saved idx={idx}  ({label})  valid={int(valid[idx])}")

# ==========================
# 5) GDS 
# ==========================
def find_gds_path(struct_idx: int) -> str | None:
    a = os.path.join(GDS_DIR, f"structure_{struct_idx:05d}.gds")
    b = os.path.join(GDS_DIR, f"structure_{struct_idx}.gds")
    if os.path.exists(a):
        return norm_path(a)
    if os.path.exists(b):
        return norm_path(b)
    return None

def gds_meta(path: str) -> dict:
    st = os.stat(path)
    return {
        "size_kb": st.st_size / 1024,
        "mtime": time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(st.st_mtime)),
    }

# ==========================
# 6) 
# ==========================
r16 = np.array(raw16[idx]).squeeze()     # (16,16)
im  = np.array(struct[idx]).squeeze()    # (128,128)
s   = np.array(spec[idx])                # (3,301)
vv  = int(valid[idx])

gds_path = find_gds_path(idx)
if gds_path is None:
    gds_info = "MISSING"
else:
    meta = gds_meta(gds_path)
    gds_info = f"{gds_path} (size={meta['size_kb']:.1f}KB, mtime={meta['mtime']})"

if PRINT_RAW16_NUMBERS:
    print(f"\n===== idx={idx} valid={vv} RAW16 (16x16) =====")
    with np.printoptions(precision=4, suppress=True, linewidth=200):
        print(r16)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].imshow(1 - r16, cmap="gray", origin="lower")
axes[0].set_title(f"RAW 16x16 (idx={idx})")
axes[0].axis("off")

axes[1].imshow(im, cmap="gray", origin="lower")
axes[1].set_title(f"STRUCT 128x128 (valid={vv})")
axes[1].axis("off")

x = np.arange(s.shape[1])
axes[2].plot(x, -s[0], label="Trans_1")
axes[2].plot(x, -s[1], label="Trans_2")
axes[2].plot(x, -s[2], label="Trans_3")
axes[2].set_title("Spectrum")
axes[2].set_xlabel("Index")
axes[2].set_ylim(-0.05, 1.05)
axes[2].grid(True, alpha=0.3)
axes[2].legend(fontsize=8)

fig.text(0.5, 0.01, f"GDS: {gds_info}", ha="center", va="bottom", fontsize=8)
plt.tight_layout()
plt.show()
```
